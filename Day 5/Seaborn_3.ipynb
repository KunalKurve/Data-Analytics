```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('StudentsPerformance.csv')
print(df.head())
```

```python
#Violin Plot

plt.figure(figsize=(12,6))
sns.violinplot(x = 'parental level of education', y = 'math score', data = df)
plt.show()
```

```python
#Violin Plot

plt.figure(figsize=(12,6))
sns.violinplot(x = 'parental level of education', y = 'math score', data = df, hue = 'parental level of education')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education', y = 'math score', data = df, hue = 'gender')
plt.show()
```

```python
# Split

plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education', y = 'math score', data = df, hue = 'gender', split = True)
plt.show()
```

Swarmplot

```python
sns.swarmplot(x='math score', data = df)
plt.show()
```

```python
sns.swarmplot(x='math score', data = df, size = 2)
plt.show()
```

```python
sns.swarmplot(x='math score', data = df, size =3, y = 'race/ethnicity')
plt.show()
```

```python
sns.swarmplot(y='math score', data = df, size =3, x = 'race/ethnicity')
plt.show()
```

Boxen Plot

```python
sns.boxenplot(x = 'math score', y = 'race/ethnicity', data = df, hue = 'race/ethnicity')
plt.show()
```

```python
sns.boxenplot(y='math score', x = 'race/ethnicity', data=df, hue = 'race/ethnicity')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxenplot(y='math score', x = 'race/ethnicity', data=df, hue = 'gender')
plt.show()
```

Heat Balst

```python
#2020 Projections
df = pd.read_csv('country_table.csv')
print(df)
```

```python
df = df.set_index('Countries')
print(df)
```

```python
sns.heatmap(df)
plt.show()
```

```python
rates = df.drop('Life expectancy', axis = 1)
sns.heatmap(rates)
plt.show()
```

```python
sns.heatmap(rates, lw = 1) #linewidth
plt.show()
```

```python
sns.heatmap(rates, lw = 0.5, annot = True) #annotation
plt.show()
```

```python
sns.heatmap(rates, lw = 0.5, annot =True, cmap = 'viridis') #colormap
plt.show()
```

```python
# Set colorbar based on value from dataset

sns.heatmap(rates, lw = 0.5, annot =True, cmap = 'viridis', center = 40) #color is basdd on the value on center
plt.show()
```
