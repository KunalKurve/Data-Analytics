```python
from sklearn.datasets import fetch_california_housing #pre-installed datasets for testing in sklearn
```

```python
housing, target = fetch_california_housing(as_frame = True, return_X_y = True)
print(housing.head())
print(target.head())
```

```python
#Correlation of all the cloumns with each other
print(housing.corr())
```

```python
#Correlation of two specific columns
corr = housing.corr()
print(corr["MedInc"]["AveRooms"])
```

```python
#Visual View
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cmap = sns.diverging_palette(10,220, as_cmap =True)
sns.heatmap(corr, vmin=-1.0, vmax = 1.0, square = True, cmap = cmap)
plt.show()
```

Exercise to find correlation between Glucose and Blood Pressure column in Diabetes dataset and also plot

```python
import pandas as pd
```

```python
diabetes = pd.read_csv("diabetes.csv")
```

```python
print(diabetes.head())
```

```python
print(diabetes.corr())
```

```python
#Correlation of two specific columns
corr = diabetes.corr()
print(corr["Glucose"]["BloodPressure"])

# correlation_coefficient = diabetes["Glucose"].corr(diabetes["BloodPressure"])
# print(correlation_coefficient)
```

```python
cmap = sns.diverging_palette(10,220, as_cmap =True)
sns.heatmap(corr, vmin=-1.0, vmax = 1.0, square = True, cmap = cmap)
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(data = diabetes, x=diabetes["Glucose"], y=diabetes["BloodPressure"]);
# plt.scatter(diabetes["Glucose"], diabetes["BloodPressure"]);
```

```python
#Box Plot
plt.figure(figsize=(12,8))
sns.boxplot(x=diabetes["Glucose"], y=diabetes["BloodPressure"])
plt.show()
```

The above graphs indicate that There is very low correlation between Glucose and Blood pressure
