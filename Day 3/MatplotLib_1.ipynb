# Matplotlib Object Oriented Method

```python
a = np.linspace(0,10,11)
b = a ** 4

print(a)
print(b)
```

```python
x = np.arange(0,10)
y = 2 * x

print(x)
print(y)
```

```python
# Creating a Figure

# The main idea in using the more formal Object Oriented method is to create a
#   figure object and then just call methods or attributes off of that object -
#   This approach is nicer when dealing with a canvas that has multiple plots on it
```

```python
# Create an empty canvas
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y)
plt.show()
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(a, b)
plt.show()
```

```python
# Adding another set of axes to the Figure
# So far we have only seen one set of axes on this figure object, 
# but we can keep adding new axes on to it at any location and size we want
# We can then plot on that new set of axes
# Code is a little more complicated, but the advantage is that we now 
# have full control of where the plot axes are placed, and we can easily 
# add more than one axis to the figure
# Note how we are plotting a, b twice here
# Create blank canvas

fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) # Large Figure
axes2 = fig.add_axes([0.2,0.2,0.5,0.5]) # Samll Figure

axes1.plot(a,b)
axes1.set_xlabel('X-axis')
axes1.set_ylabel('Y-axis')
axes1.set_title('Big Figure')

axes2.plot(a,b)
axes2.set_title('Small Figure')
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) # Large Figure
axes2 = fig.add_axes([0.2,0.5,0.25,0.25]) # Samll Figure

# Larger fig
axes1.plot(a,b)

axes1.set_xlabel('X-axis')
axes1.set_ylabel('Y-axis')
axes1.set_title('Big Figure')

# Insert fig axes 2

axes2.plot(a,b)


axes2.set_xlim(0,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed in')
plt.show()
```

```python
x = np.arange(0,10)
y = 2 * x

print(x)
print(y)
```

```python
# The plt.subplots() object will act as a more automatic axis manager. 
# This make sit much easier to show multiple plots side by side

figure,axes = plt.subplots()

# now use the axes object to add stuff to the plot

axes.plot(x,y,'r')  # r= red color
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')  # hides out[]
plt.show()
```

```python
# Empty canvas of 1 by 2 sub plots

figure,axes = plt.subplots(nrows=1, ncols=2)
print(axes)
print(axes.shape)
plt.show()
```

```python
# Empty canvas of 2 by 2 sub plots

figure,axes = plt.subplots(nrows=2, ncols=2)
print(axes)
print(axes.shape)
plt.show()
```

```python
# Use .subplots_adjust to adjust spacing manually

figure,axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
# This is a standard figure size

# Parameter at the axes level
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

# Use left, right, top ,bottom to stretch subplots
# Use wspace, hspace to add spacing between subplots

figure.subplots_adjust(left = None, bottom = None, right = None, top = None, wspace = 0.9, hspace = 0.2)
plt.show()
```

```python
x = np.arange(0,10)
y = x * 2

# Legends
# We can use the **label = "Label Text"** keyword argument when plots or other objects are added to the figure
# and then using the **legend** method without arguments to add the legend to the figure

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label = "x**2")
ax.plot(x, x**3, label = "x**3")
ax.legend(loc=1)  # Upper right corner
plt.show()
```

```python
# Setting color, line width, line types
# Matlab style line color and style

fig, ax = plt.subplots()

ax.plot(x, x**2, 'b.-') # Blue line with dots
ax.plot(x, x**3, 'g--')  # Green dashed line

plt.show()
```

```python
# Suggested Approach: Use keyword arguments
fig, ax = plt.subplots()

ax.plot(x, x+1, color = 'blue', alpha = 0.5) # 
ax.plot(x, x+2, color = "#88008B")  # 
ax.plot(x, x+3, color = "#FF8C00")  #

plt.show()
```

```python
# Line and Marker Style

fig, ax = plt.subplots()

ax.plot(x, x+1, color = 'red', linewidth = 0.25) # 
ax.plot(x, x+2, color = 'red', lw = 0.5)  # 
ax.plot(x, x+3, color = 'red', lw = 1)  #
ax.plot(x, x+4, color = 'red', lw = 10)  #


plt.show()
```

```python
# Linestyles

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color = 'blue', lw = 3, linestyle = '-') # solid
ax.plot(x, x+2, color = 'blue', lw = 3, ls = '-.')  # Dash and dot
ax.plot(x, x+3, color = 'blue', lw = 3, ls = ':')  # Dots
ax.plot(x, x+4, color = 'blue', lw = 3, ls = '--')  # Dashes


plt.show()
```

```python
# Markers

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, marker = '+', markersize = 20) 
ax.plot(x, x+2, marker = 'o', ms = 20)  
ax.plot(x, x+3, marker = 's', ms = 20 , lw = 0)  # make line width = 0 to see only markers
ax.plot(x, x+4, marker = '1', ms = 20)  # 


plt.show()
```
