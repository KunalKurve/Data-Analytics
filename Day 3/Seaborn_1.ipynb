```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("dm_office_sales.csv")

print(df.head())
print(df.info())
```

```python
sns.scatterplot(x='salary', y='sales', data = df)
plt.show() #Note that seaborn is linked to matplotlib underneath
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales', data = df)
plt.show()
```

```python
#hue: Color points based off a categorical feature in the dataFrame

plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales', data = df, hue='division')
plt.show()
```

```python
#hue: Color points based off a categorical feature in the dataFrame
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales', data = df, hue='work experience')
plt.show()
```

```python
# Choosing a palette from Matplotlib's cmap
# https://matplotlib.org/tutorials/colors/colormaps.html
 
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales', data = df, hue='work experience', palette = 'viridis')  # Most used professionally
plt.show()
```

```python
# Use *s=* if you want to change the marker size to b some uniform integer value

plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales', data = df, hue='work experience', palette = 'viridis', s=200)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales', data = df, s = 200, lw = 0, alpha = 0.2)
plt.show()
```

```python
# Style
# Automatically choose styles based on another categorical feature in the dataset
#   Optionally use the **markers=** parameter to pass a list of marker choices 
#   based off matplotlib, for example: ['*','+','o']

plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales', data = df,style = "level of education")
plt.show()
```

```python
# Sometimes it is nice to do both hue and style off the same column

plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales', data = df, style = "level of education", hue = 'level of education', s = 100)
plt.show()
```

# Distribution Plots

```python
# Rug Plot

sns.rugplot(x = 'salary', data = df)
sns.rugplot(x = 'salary', data = df, height = 0.5)
plt.show()
```

```python
# displot: Distribution of a single feature
sns.displot(x = 'salary', data = df, kde = False)   
plt.show()
```

```python
sns.displot(x = 'salary', data = df, kde = True)
#kde: Kernel Density: Shows the cumulative value of the distributive plot. 
# The curve represents KDE
plt.show()
```

Histogram Plots

```python
# Bins
sns.histplot(x = 'salary', data = df, bins = 10)
plt.show()
```

```python
# Bins
sns.histplot(x = 'salary', data = df, bins = 100)
plt.show()
```

KDE(Kernel Density Estimation) Plot


```python
np.random.seed(42)

sample_ages = np.random.randint(0,100,200)
print(sample_ages)
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
print(sample_ages.head())
```

```python
sns.rugplot(data = sample_ages, x = 'age')
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.displot(data = sample_ages, x = 'age', bins = 10, rug = True)
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.displot(data = sample_ages, x = 'age', bins = 10, rug = True, kde = True)
plt.show()
```

```python
sns.kdeplot(data = sample_ages, x = 'age')
plt.show()
```
