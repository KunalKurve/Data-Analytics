```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("dm_office_sales.csv")

print(df.head())
print(df.info())
```

```python
plt.figure(figsize = (12,8))
sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = np.mean, errorbar = 'sd')
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = np.mean)
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = np.mean, errorbar = 'sd', hue = 'division')
plt.show()
```

```python
df = pd.read_csv('StudentsPerformance.csv')
print(df.head())
```

```python
#box Plot
#display distribution through the use of quartiles and an IQR for ourliesrs
plt.figure(figsize=(12,6))
sns.boxplot(x = 'parental level of education', y = 'math score', data = df)
plt.show()
```

```python
#box Plot
#display distribution through the use of quartiles anf an IQR for ourliesrs
plt.figure(figsize=(12,6))
sns.boxplot(x = 'parental level of education', y = 'math score', data = df, hue = 'parental level of education')
plt.show()
```

```python
# Changing Orientation

plt.figure(figsize=(12,6))
sns.boxplot(y = 'parental level of education', x = 'math score', data = df, orient = 'h', hue = 'gender')
plt.show()
```

```python
#Violin Plot

plt.figure(figsize=(12,6))
sns.violinplot(x = 'parental level of education', y = 'math score', data = df)
plt.show()
```
