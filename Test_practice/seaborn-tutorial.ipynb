# <span style="color: #007BFF; font-size: 24px; font-weight: bold; border: 2px solid red; padding: 10px; background-color: #EFEFEF; display: block; width: 100%;">➡️ 2. Distribution plot</span>

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <p>
        Distribution of data is the foremost thing that we need to understand while analysing the data. Here, we will see how seaborn helps us in understanding the distribution of our data.
   </p>

   <span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid; padding: 10px;  display: block; width: 100%;">
        2.1. distplot
    </span>

   <p>
       The <code>distplot()</code> function provides the most convenient way to take a quick look at univariate distribution. This function will plot a <em>histogram</em> that fits the <strong>kernel density estimation(KDE)</strong> of the data.
    </p>

   <p>
        Now let's plot the histogram of <strong>Number of parents and children of the passenger aboard(parch)</strong>.
    </p>

</div>


```python
sns.histplot(full_data['Parch'],kde=False)
plt.show()
```

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">
   <p>
        As we can see, most passengers don't have neither parents nor children aboard.
    </p>

   <span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid; padding: 10px;  display: block; width: 100%;">
        2.2. kdeplot
    </span>

   <p>
        <em>Kernel Density Estimation (KDE)</em> is a way to estimate the probability density function of a continuous random variable. It is used for <strong>non-parametric</strong> analysis. Setting the <code>hist</code> flag to False in <code>distplot</code> will yield the KDE plot. For bivariate distribution, we can plot a kde by using <code>jointplot()</code>. Pass value <code>‘kde’</code> to the parameter <code>kind</code> to plot the kernel plot.
    </p>

   <p>
        <strong>Note:</strong> <code>distplot(data)</code> is used to visualize the <em>parametric distribution</em> of data. It plots both KDE and histogram on the same figure.
    </p>
</div>


```python
sns.distplot(full_data['Age'], hist=False)
plt.show()
```

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

As we can see that most of the passenger has the age between 20 to 40
    </div>

```python
plt.figure(figsize=(8,8))
sns.distplot(full_data['Age'])
plt.show()
```

# <span style="color: #007BFF; font-size: 24px; font-weight: bold; border: 2px solid red; padding: 10px; background-color: #EFEFEF; display: block; width: 100%;">➡️ 3. Relational plots</span>

<br><span style="color: #007BFF; font-size: 24px; font-weight: bold; border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    3.1. relplot
</span>
<br><br>
<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <p>
        Figure-level interface for drawing relational plots onto a FacetGrid.
    </p>

   <p>
        The function <code>relplot()</code> is named that way because it is designed to visualize many different statistical relationships. While scatter plots are a highly effective way of doing this, relationships where one variable represents a measure of time are better represented by a line. The <code>relplot()</code> function has a convenient <code>kind</code> parameter to let you easily switch to this alternate representation.
    </p>

</div>


```python
sns.relplot(x="Age", y="Fare", col="Pclass", hue="Sex", style="Sex",kind="line", data=full_data) # scatter can be used instead of "line" plot
plt.show()
```

<span style="color: #007BFF; font-size: 24px; font-weight: bold; border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    3.2. scatterplot
</span>
<br><br>


<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <p>
        <strong>Scatter plot</strong> is the most convenient way to visualize the distribution where each observation is represented in a two-dimensional plot via x and y axis.
    </p>

</div>


```python
plt.figure(figsize=(8,8))
sns.scatterplot(x="Age", y="Fare", hue="Sex", data=full_data)
plt.show()
```

<span style="color: #007BFF; font-size: 24px; font-weight: bold; border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    3.3. lineplot
</span>
<br><br>

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <p>
        Draw a line plot with the possibility of several semantic groupings.
    </p>

   <p>
        The relationship between <code>x</code> and <code>y</code> can be shown for different subsets of the data using the <code>hue, size</code>, and <code>style</code> parameters. These parameters control what visual semantics are used to identify the different subsets. It is possible to show up to three dimensions independently by using all three semantic types, but this style of plot can be hard to interpret and is often ineffective. Using redundant semantics (i.e., both <strong>hue</strong> and <strong>style</strong> for the same variable) can be helpful for making graphics more accessible.
    </p>

   <p>
        The default treatment of the hue (and to a lesser extent, size) semantic, if present, depends on whether the variable is inferred to represent <strong>“numeric” or “categorical”</strong> data. In particular, numeric variables are represented with a sequential colormap by default, and the legend entries show regular <strong>“ticks”</strong> with values that may or may not exist in the data. This behavior can be controlled through various parameters.
    </p>

   <p>
        By default, the plot aggregates over multiple <code>y</code> values at each value of <code>x</code> and shows an estimate of the central tendency and a confidence interval for that estimate.
    </p>

</div>


```python
plt.figure(figsize=(8,8))
sns.lineplot(x="Age", y="Fare", hue="Sex", style="Sex", data=full_data)
plt.show()
```

# <span style="color: #007BFF; font-size: 24px; font-weight: bold; border: 2px solid #007BFF; padding: 10px; background-color: #EFEFEF; display: block; width: 100%;">➡️ 4. Categorical Plot</span>

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <p>
        When one or both the variables under study are categorical, we use plots like <code>striplot(), swarmplot(), etc,</code>. Seaborn provides an interface to do so.
    </p>

   <span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
        4.1. barplot
    </span>
   <br><br>

   <p>
        <em>The <code>barplot()</code> shows the relation between a categorical variable and a continuous variable.</em> The data is represented in rectangular bars where the length of the bar represents the proportion of the data in that category.
        <strong>Bar plot</strong> represents the estimate of <em>central tendency.</em>
    </p>

   <p>
        <strong>Note:</strong> Don't confuse <strong>Bar plot</strong> and <strong>Histogram</strong>. Please go back to <code>2.1. distplot</code> section to see the difference.
    </p>

</div>


```python
plt.figure(figsize=(8,8))
sns.barplot(x="Sex", y="Survived", hue="Pclass", data=full_data)
plt.show()
```

**As we can see, More women survived than men.**


<span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    4.2. stripplot
</span>
<br><br>

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <p>
        <code>stripplot()</code> is used when one of the variables under study is categorical. It represents the data in sorted order along any one of the axes.
    </p>

</div>


```python
plt.figure(figsize=(8,8))
sns.stripplot(x="Sex", y="Age",hue='Sex', data=full_data)
plt.show()
```

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <p>
        To avoid the overlapping of the points, we can use the <code>jitter</code> to add some random noise to the data. This parameter will adjust the positions along the categorical axis. But another option which can be used as an alternate to <code>‘Jitter’</code> is the function <code>swarmplot()</code>.
    </p>
   <br>

   <span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
        4.3. swarmplot
    </span>
   <br><br>

   <p>
        This function positions each point of the scatter plot on the categorical axis and thereby avoids overlapping points:
    </p>

</div>


```python
plt.figure(figsize=(8,8))
sns.swarmplot(x="Sex", y="Age",hue='Sex', data=full_data)
plt.show()
```

<strong>
We can said that more passengers are approximally between 18 and 40 years old.</strong>

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <p>
        <span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
            4.4. boxplot
        </span>
       <br><br>
       <strong>Boxplot</strong> is a convenient way to visualize the distribution of data through their quartiles. Box plots usually have vertical lines extending from the boxes which are termed as <strong>whiskers</strong>. These whiskers indicate variability outside the upper and lower quartiles, hence Box Plots are also termed as <strong>box-and-whisker plot</strong> and <strong>box-and-whisker diagram</strong>. Any Outliers in the data are plotted as individual points.
    </p>

</div>


```python
plt.figure(figsize=(8,8))
sns.boxplot(x="Survived", y="Age", data=full_data)
plt.show()
```

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <p>
        <strong>We have some outliers for passengers surveyed through their age.</strong>
    </p>

   <span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
        4.5. violinplot
    </span>
   <br><br>

   <p>
        <strong>Violin Plots</strong> are a combination of the <strong>box plot</strong> with the <strong>kernel density estimates</strong>. So, these plots are easier to analyze and understand the distribution of the data.
    </p>

</div>


```python
sns.violinplot(x="Survived", y="Age", hue='Sex', data=full_data)
plt.show()
```

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <p>
        The quartile and whisker values from the boxplot are shown inside the violin. As the violin plot uses KDE, the wider portion of the violin indicates higher density, and the narrow region represents relatively lower density. The Inter-Quartile range in boxplot and higher density portion in KDE fall in the same region of each category of the violin plot.
    </p>

   <span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
        4.6. countplot
    </span>
   <br><br>

   <p>
        A special case in <strong>barplot</strong> is to show the number of observations in each category rather than computing a statistic for a second variable. For this, we use <code>countplot()</code>.
    </p>

</div>


```python
sns.countplot(x="Survived", data=full_data, palette="Blues");
plt.show()
```

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

   <span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
        4.7. pointplot
    </span>
   <br><br>

   <p>
        <strong>Point plots</strong> serve the same as <strong>bar plots</strong> but in a different style. Rather than the full bar, the value of the estimate is represented by the point at a certain height on the other axis.
    </p>

</div>


```python
plt.subplots(figsize=(8, 8))
sns.pointplot(x="Sex", y="Survived", hue="Pclass", data=full_data)
plt.show()
```

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

<strong>We can see the average number of survivals of male and female in each class. From the plot we can understand that more number of females survived than males. In both males and females more number of survivals are from first
class.</strong>
    
</div>


<strong>Plot says that the number of passengers in the third class is higher than the first and second class.</strong>

# <span style="color: #007BFF; font-size: 24px; font-weight: bold; border: 2px solid #007BFF; padding: 10px; background-color: #EFEFEF; display: block; width: 100%;">➡️ 7. Multi-plot grids</span>
<br>

<span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    7.1. Facet grids
</span>
<br><br>

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">
   <p>
    A useful approach to explore medium-dimensional data is by drawing multiple instances of the same plot on different subsets of your dataset. This technique is commonly called <em>“lattice”</em> or <em>“trellis”</em> plotting, and it is related to the idea of <em>“small multiples”</em>. To use these features, your data has to be in a Pandas DataFrame.
 </p>

   <p>
        <strong>Facet grid</strong> forms a matrix of panels defined by row and column by dividing the variables. Due to panels, a single plot looks like multiple plots. It is very helpful to analyze all combinations in two discrete variables.
    </p>
   <p>
        <strong>FacetGrid</strong> object takes a dataframe as input and the names of the variables that will form the row, column, or hue dimensions of the grid. The variables should be categorical, and the data at each level of the variable will be used for a facet along that axis.
    </p>
   <p>
        The advantage of using Facet is we can input another variable into the plot. We can make many column facets and align them with the rows of the grid.
    </p>
</div>


#### FacetGrid.map
The main approach for visualizing data on this grid is with the `FacetGrid.map()` method.

```python
# initialize the FacetGrid object
g = sns.FacetGrid(full_data, col='Survived', row='Pclass')

g.map(plt.hist, 'Age')
g.add_legend()
plt.show()
```
