```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("dm_office_sales.csv")
```

```python
print(df.head())
print(df.info())
```

```python
sns.scatterplot(x='salary',y='sales',data=df)
plt.show() # Note that seaborn is linked with matplotlib underneath
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df)
plt.show()
```

```python
#hue: Color points based off a categorical feature in the Data frame
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='division')
plt.show()
```

```python
#hue: Color points based off a categorical feature in the Data frame
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience')
plt.show()
```

```python
#Chossing a palette from Matplotlib's cmap:
# https://matplotlib.org/tutorials/color/colormaps.html
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='viridis')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='plasma')
plt.show()
```

```python
#S "s=" if you want to change the marker size to be som uniform integer value
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='viridis',s=100)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='viridis',s=100,linewidth=0,alpha=0.2)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',palette='plasma')
plt.show()
```

```python
sns.rugplot(x='salary',data=df)
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
#displot: Distribution of a single feature
sns.set(style='dark')
sns.displot(data=df,x='salary',kde=False,color='black');
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary',kde=True,color='gold');
```

```python
sns.set(style='dark')
sns.displot(data=df,x='salary',kde=True,color='blue',bins=10);
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary',kde=True,color='gold',bins=100);
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary',kde=True,color='red',bins=100);
```

```python
sns.set(style='white')
sns.displot(data=df,x='salary',kde=True,color='green',bins=100);
```

```python
np.random.seed(42)
sample_ages=np.random.randint(0,100,200)
print(sample_ages)
```

```python
sample_ages=pd.DataFrame(sample_ages,columns=['age'])
print(sample_ages.head())
```

```python
sns.rugplot(data=sample_ages,x='age')
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages,x='age',bins=10,rug=True)
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages,x='age',bins=10,rug=True,kde=True)
```

```python
plt.figure(figsize=(12,8))
sns.kdeplot(data=sample_ages,x='age')
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,errorbar='sd',hue='division',palette='plasma')
```

```python
df=pd.read_csv('StudentsPerformance.csv')
print(df.head)
```

```python
#box Plot
#Display distribution through the use of quartiles and an IQR for outliers
plt.figure(figsize=(12,8))
sns.boxplot(x='parental level of education',y='math score',data=df)
```

```python
#box Plot
#Display distribution through the use of quartiles and an IQR for outliers
plt.figure(figsize=(12,8))
sns.boxplot(x='parental level of education',y='math score',data=df,hue='gender',palette='plasma')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(y='parental level of education',x='math score',data=df,orient='h')
```
