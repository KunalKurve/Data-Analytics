```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,10)
y=np.arange(1,19)

y=y[:9]

plt.plot(x,y)

plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Plot Title Here')
plt.scatter(x,y,s=50)
```

```python
a= np.linspace(0, 11, 10)
b = a ** 4
print(a)
print(b)
```

```python
#Create an empty figure
fig=plt.figure()
#Add set of axes to the figure
axes=fig.add_axes([0,0,1,1]) # left,bottom,width,height (range 0 to 1)
#Plot on that set of axes
axes.plot(x,y)
```

```python
#Create an empty figure
fig=plt.figure()
#Add set of axes to the figure
axes=fig.add_axes([0,0,1,1]) # left,bottom,width,height (range 0 to 1)
#Plot on that set of axes
axes.plot(a,b)
```

```python
#Create an empty figure
fig=plt.figure()
#Add set of axes to the figure
axes1 =fig.add_axes([0,0,1,1]) #left,bottom,width,height (range 0 to 1)
axes2 =fig.add_axes([0.2,0.2,0.5,0.5])
#Larger Figure Axes
axes1.plot(a,b)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Big Figure');
#Smaller Figure Axes
axes2.plot(a,b)
plt.title('Small Figure');
```

```python
#Create an empty figure
fig=plt.figure()
#Add set of axes to the figure
axes1 =fig.add_axes([0,0,1,1]) #left,bottom,width,height (range 0 to 1)
axes2 =fig.add_axes([0.2,0.5,0.25,0.25])
#Larger Figure Axes
axes1.plot(a,b)
axes1.set_xlabel('X Axis Title Here')
axes1.set_ylabel('Y Axis Title Here')
axes1.set_title('Big Figure');
#Smaller Figure Axes
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed in');
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(b,a)
axes[1][0].plot(y,x)

fig.subplots_adjust(left=None,bottom=None,right=None,top=None,wspace=0.9,hspace=0.1,)
plt.show()
```

```python
x=np.arange(0,10)
y=2*x

fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="x**2")
ax.plot(x,x**3, label="x**3")
ax.set_title('LEGENDS')
ax.legend(loc=0);
```

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,'b.-')
ax.plot(x,x**3,'g--')
```

```python
fig, ax = plt.subplots()
ax.plot(x,x+1,color='#ffb3ba',alpha=0.5)
ax.plot(x,x+2,color='#CCFF00')
ax.plot(x,x+3,color='#512467')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x,x+1,marker='+',ms=20)
ax.plot(x,x+2,marker='o',ms=20)
ax.plot(x,x+3,marker='s',ms=20,lw=0)
ax.plot(x,x+4,marker='1',ms=20)
```
