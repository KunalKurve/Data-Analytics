```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("LS2009Candidate.csv")
df1.head(10)
```

```python
df2 = pd.read_csv("LS2009Electors.csv")
df2.head()
```

```python
df3 = pd.read_csv("LS2014Candidate.csv")
df3.head(10)
```

```python
df4 = pd.read_csv("LS2014Electors.csv")
df4.head()
```

```python

nda_parties = ['BJP','SS', 'LJP', 'SAD', 'RLSP','AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF','NDPP','MNF','RIDALOS','KMDK','IJK','PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP','KPP','JKPC','KC(T)','BDJS','AGP','JSS','PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP','KC(N)','PDF','MDPF']
upa_parties = ['INC','NCP', 'RJD', 'DMK', 'IUML','JMM','JD(s)','KC(M)','RLD','RSP','CMP(J)','KC(J)','PPI','MD']

def assign_alliance(party):
    if party in nda_parties:
        return 'NDA'
    elif party in upa_parties:
        return 'UPA'
    else:
        return 'Others'

df1['Alliance'] = df1['Party Abbreviation'].apply(assign_alliance)
df3['Alliance'] = df3['Party Abbreviation'].apply(assign_alliance)

df = pd.concat([df1, df3])
df5 = pd.concat([df2, df4])

df1.head()
df5.head()
```

```python
df_grouped=df1.groupby(df1['Alliance'])
df_grouped.value_counts()
```

```python
# 2. Create Winning seats distribution by Major Political Parties & Alliances for 2009 & 2014

winners = df[df['Position'] == 1]

seats_won_2009 = winners[winners['Year'] == 2009]['Alliance'].value_counts()
seats_won_2014 = winners[winners['Year'] == 2014]['Alliance'].value_counts()

seat_counts = pd.DataFrame({'2009': seats_won_2009, '2014': seats_won_2014})

seat_counts.plot(kind='bar', stacked=False, figsize=(10,6))

plt.title('Winning Seats Distribution - 2009 & 2014')
plt.ylabel('No. of Seats Won')
plt.xlabel('Alliance')
plt.show()
```

```python
# 3. How many seats won by Alliances and Major Political Parties ?

winners = df[df['Position'] == 1]

# Count no. of seats won by each alliance
seats_won_by_alliance = winners['Alliance'].value_counts()

# Count no. of seats won by each party
seats_won_by_party = winners['Party Abbreviation'].value_counts()

print("Seats won by each alliance:")
print(seats_won_by_alliance)
print("\nSeats won by each party:")
print(seats_won_by_party)
```

```python
# 4. Plot comparatively seats won based on candidate category as General, ST and SC for 2009 & 2014

winners = df[df['Position'] == 1]

# Count no. of seats won by each category in 2009 and 2014
seats_won_2009 = winners[winners['Year'] == 2009]['Candidate Category'].value_counts()
seats_won_2014 = winners[winners['Year'] == 2014]['Candidate Category'].value_counts()

# Counting seats
seat_counts = pd.DataFrame({'2009': seats_won_2009, '2014': seats_won_2014})

# Plotting Bar Plot
seat_counts.plot(kind='bar', figsize=(10,6))

plt.title('Seats won based on candidate category as General, ST and SC for 2009 & 2014')
plt.ylabel('Number of Seats Won')
plt.xlabel('Candidate Category')
plt.show()
```

```python
# 5. Plot the age distribution of winners of both 2014 and 2009 elections

winners = df[df['Position'] == 1]

# Winners for 2009 and 2014
winners_2009 = winners[winners['Year'] == 2009]['Candidate Age']
winners_2014 = winners[winners['Year'] == 2014]['Candidate Age']

# Plotting Histogram
plt.figure(figsize=(10,6))
sns.histplot(winners_2009, color='blue', kde=False, label='2009')
sns.histplot(winners_2014, color='red', kde=False, label='2014')

plt.title('Age Distribution of Winners of 2014 and 2009 Elections')
plt.xlabel('Candidate Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# 6. Separately plot age distribution of NDA & UPA candidates

nda_candidates = df[df['Alliance'] == 'NDA']['Candidate Age']
upa_candidates = df[df['Alliance'] == 'UPA']['Candidate Age']

# Plotting Histogram of NDA candidates
plt.figure(figsize=(10,6))
sns.histplot(nda_candidates, color='blue', kde=False)
plt.title('Age Distribution of NDA Candidates')
plt.xlabel('Candidate Age')
plt.ylabel('Frequency')
plt.show()

# Plotting Histogram of UPA candidates
plt.figure(figsize=(10,6))
sns.histplot(upa_candidates, color='Green', kde=False)
plt.title('Age Distribution of UPA Candidates')
plt.xlabel('Candidate Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 7. Plot Gender distributions of 2009 & 2014 elections

candidates_2009 = df[df['Year'] == 2009]['Candidate Sex'].value_counts()
candidates_2014 = df[df['Year'] == 2014]['Candidate Sex'].value_counts()

# Counting Genders
gender_counts = pd.DataFrame({'2009': candidates_2009, '2014': candidates_2014})

# Plotting Bar Plot
gender_counts.plot(kind='bar', figsize=(10,6))

plt.title('Gender Distribution of Candidates in 2009 & 2014 Elections')
plt.ylabel('Number of Candidates')
plt.xlabel('Gender')
plt.show()
```

```python
# 8. Plot gender distribution of NDA and UPA separately for 2009 elections

nda_candidates_2009 = df[(df['Year'] == 2009) & (df['Alliance'] == 'NDA')]['Candidate Sex'].value_counts()
upa_candidates_2009 = df[(df['Year'] == 2009) & (df['Alliance'] == 'UPA')]['Candidate Sex'].value_counts()

# Plotting Bar Plot for NDA candidates
plt.figure(figsize=(10,6))
sns.barplot(x=nda_candidates_2009.index, y=nda_candidates_2009.values, alpha=0.8)
plt.title('Gender Distribution of NDA Candidates in 2009 Elections')
plt.ylabel('Number of Candidates')
plt.xlabel('Gender')
plt.show()

# Plotting Bar Plot for UPA candidates
plt.figure(figsize=(10,6))
sns.barplot(x=upa_candidates_2009.index, y=upa_candidates_2009.values, alpha=0.8)
plt.title('Gender Distribution of UPA Candidates in 2009 Elections')
plt.ylabel('Number of Candidates')
plt.xlabel('Gender')
plt.show()
```

```python
# 9. Plot gender distribution of NDA and UPA separately for 2014 elections

nda_candidates_2014 = df[(df['Year'] == 2014) & (df['Alliance'] == 'NDA')]['Candidate Sex'].value_counts()
upa_candidates_2014 = df[(df['Year'] == 2014) & (df['Alliance'] == 'UPA')]['Candidate Sex'].value_counts()

# Plotting Bar Plot for NDA candidates
plt.figure(figsize=(10,6))
sns.barplot(x=nda_candidates_2014.index, y=nda_candidates_2014.values, alpha=0.8)
plt.title('Gender Distribution of NDA Candidates in 2009 Elections')
plt.ylabel('Number of Candidates')
plt.xlabel('Gender')
plt.show()

# Plotting Bar Plot for UPA candidates
plt.figure(figsize=(10,6))
sns.barplot(x=upa_candidates_2014.index, y=upa_candidates_2014.values, alpha=0.8)
plt.title('Gender Distribution of UPA Candidates in 2009 Elections')
plt.ylabel('Number of Candidates')
plt.xlabel('Gender')
plt.show()
```

```python
# 10. Plot the poll percentage of states for 2009 & 2014 elections

plt.figure(figsize=(15,8))
plt.bar(x=df2['STATE'], height=df2['POLL PERCENTAGE'], width=0.4, label='2009', align='center')
plt.bar(x=df4['STATE'], height=df4['POLL PERCENTAGE'], width=0.4, label='2014', align='edge')
plt.xlabel('State name')
plt.ylabel('Poll Percentage')
plt.title('Poll Percentage of States for 2009 & 2014 Elections')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
