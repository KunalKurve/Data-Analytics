```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("LS2009Candidate.csv")
df1
```

```python
df2 = pd.read_csv("LS2009Electors.csv")
df2
```

```python
df3 = pd.read_csv("LS2014Candidate.csv")
df3
```

```python
df4 = pd.read_csv("LS2014Electors.csv")
df4
```

```python
#Concatenating Candidate Data
dfC = pd.concat([df1, df3], axis = 0)
dfC
```

```python
# 1. Create grand alliances
UPA = ['INC','NCP', 'RJD', 'DMK', 'IUML','JMM','JD(s)','KC(M)','RLD','RSP','CMP(J)','KC(J)','PPI','MD']
NDA = ['BJP','SS', 'LJP', 'SAD', 'RLSP','AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF','NDPP','MNF','RIDALOS','KMDK','IJK','PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP','KPP','JKPC','KC(T)','BDJS','AGP','JSS','PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP','KC(N)','PDF','MDPF']
Others = ['YSRCP','AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 'CPI', 'AIMIM','KEC(M)','SWP', 'NPEP', 'JKN', 'AIFB', 'MUL', 'AUDF', 'BOPF', 'BVA', 'HJCBL','JVM','MDMK']
def myalliance(party):
    if party in UPA:
        return 'UPA'
    elif party in NDA:
        return 'NDA'
    else:
        return 'Others'
df1['alliance'] = df1['Party Abbreviation'].apply(myalliance)

df3['alliance'] = df3['Party Abbreviation'].apply(myalliance)
dfC['alliance'] = dfC['Party Abbreviation'].apply(myalliance)
```

```python
# 2. Create Winning seats distribution by Major Political Parties & Alliances for 2009 & 2014

#Year 2009
vc = df1['alliance'].value_counts()
print(vc)
plt.pie(vc, labels=['Others', 'NDA', 'UPA'], autopct='%1.1f%%')
plt.show()
```

```python
#Year 2014
vc2 = df3['alliance'].value_counts()
print(vc2)
plt.pie(vc2, labels=['Others', 'NDA', 'UPA'], autopct='%1.1f%%')
plt.show()
```

```python
#3. How many seats won by Alliances and Major Political Parties ?

vc1 = df1['alliance'].value_counts()
print(vc1)

vc3 = df3['alliance'].value_counts()
print(vc3)
```

```python
# 4. Plot comparatively seats won based on candidate category as General, ST and SC for 2009 & 2014

#year 2009
won_09 = df1[df1['Position'] == 1]
won_09
```

```python
#Year 2009
category = ['General', 'ST' , 'SC']
cat_vc_09 = won_09['Candidate Category'].value_counts()
print(cat_vc_09)

plt.pie(cat_vc_09, labels=category, autopct='%1.1f%%')
plt.show()
```

```python
#year 2014
won_14 = df3[df3['Position'] == 1]
won_14
```

```python
#year 2014
cat_vc_14 = won_14['Candidate Category'].value_counts()
print(cat_vc_14)
plt.pie(cat_vc_14, labels = ['General', 'SC' , 'ST'], autopct='%1.1f%%')
plt.show()
```

```python
# 5. Plot the age distribution of winners of both 2014 and 2009 elections

#year 2009
# Plot the histogram of age distribution
plt.hist(won_09['Candidate Age'], bins=20, align='left', alpha=0.5, label='Number of Individuals')

plt.xlabel('Candidate Age')
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
# year 2014

# Plot the histogram of age distribution
plt.hist(won_14['Candidate Age'], bins=range(0, 101, 5), align='left', alpha=0.5, label='Number of Individuals')

plt.xlabel('Candidate Age')
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
# 6. Separately plot age distribution of NDA & UPA candidates
nda_data = dfC[dfC['alliance'] == 'NDA']

plt.hist(nda_data['Candidate Age'], bins=range(0, 101, 5), alpha=0.5, label='NDA Candidates')
plt.xlabel('Candidate Age')
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
# Plot age distribution for UPA candidates
upa_data = dfC[dfC['alliance'] == 'UPA']
plt.hist(upa_data['Candidate Age'], bins=range(0, 101, 5), alpha=0.5, label='UPA Candidates')

plt.xlabel('Candidate Age')
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
# 7. Plot Gender distributions of 2009 & 2014 elections

#year 2009
gender_data = dfC.groupby(['Candidate Sex','Year']).size().unstack()

gender_data.plot(kind = 'bar')
plt.xlabel('Candidate Sex')
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
# 8. Plot gender distribution of NDA and UPA separately for 2009 elections
gender_dist_09 = won_09.groupby(['alliance', 'Candidate Sex']).size().unstack()

# Plot the gender distribution for NDA and UPA in 2014
gender_dist_09.plot(kind='bar')
plt.title('Gender Distribution of NDA & UPA Candidates (2009)')
plt.xlabel('Alliance')
plt.ylabel('Number of Candidates')
plt.xticks(rotation=0)
plt.legend(title='Sex')
plt.tight_layout()
plt.show()
```

```python
# 9. Plot gender distribution of NDA and UPA separately for 2014 elections
gender_dist_14 = won_14.groupby(['alliance', 'Candidate Sex']).size().unstack()

# Plot the gender distribution for NDA and UPA in 2014
gender_dist_14.plot(kind='bar')
plt.title('Gender Distribution of NDA & UPA Candidates (2014)')
plt.xlabel('Alliance')
plt.ylabel('Number of Candidates')
plt.xticks(rotation=0)
plt.legend(title='Sex')
plt.tight_layout()
plt.show()
```

```python
# 10. Plot the poll percentage of states for 2009 & 2014 elections
plt.figure(figsize=(15,8))
plt.bar(df2['STATE'], df2['POLL PERCENTAGE'], width=0.4, label='2009', align='center')
plt.bar(df4['STATE'], df4['POLL PERCENTAGE'], width=0.4, label='2014', align='edge')
plt.xlabel('State name')
plt.ylabel('Poll Percentage')
plt.title('Poll Percentage of States for 2009 & 2014 Elections')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
