```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt

# a. Create a figure object and axis of the figure canvas at axis1 = [0,0,1,1]
fig = plt.figure()
axis1 = fig.add_axes([0, 0, 1, 1])
axis1.plot(x, y)
axis1.set_xlabel('x')
axis1.set_ylabel('y')
axis1.set_title('Plot of y = 2x')
```

```python
# b. Create another plot at axis1 = [0,0,1,1] and axis2 = [0.2,0.5,0.2,0.2]
fig = plt.figure()
axis1 = fig.add_axes([0, 0, 1, 1])
axis2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
axis1.plot(x, y, 'r')
axis1.set_xlabel('x')
axis1.set_ylabel('y')
axis1.set_title('Plot of y = 2x')
axis2.plot(x, z, 'b')
axis2.set_xlabel('x')
axis2.set_ylabel('z')
axis2.set_title('Plot of z = x^2')
```

```python
# c. Create a plot by adding two axis object at [0,0,1,1] & [0.2,0.5,0.4,0.4]
fig = plt.figure()
axis1 = fig.add_axes([0, 0, 1, 1])
axis2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
axis1.plot(x, y, 'r')
axis1.set_xlabel('x')
axis1.set_ylabel('y')
axis1.set_title('Plot of y = 2x')
axis2.plot(x, z, 'b')
axis2.set_xlabel('x')
axis2.set_ylabel('z')
axis2.set_title('Plot of z = x^2')
```

```python

# d. Use x, y, z array to create the plots
# i. Between x and z arrays
plt.figure()
plt.plot(x, z)
plt.xlabel('x')
plt.ylabel('z')
plt.title('Plot of z = x^2')
```

```python
# ii. Between x & y arrays
plt.figure()
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = 2x')

plt.show()
```
