```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0, 0, 1, 1])

axis1.set_xlabel("X-Axis", fontsize = 12)
axis1.set_ylabel("Y-Axis", fontsize = 12)
axis1.set_title("Plot 1", fontsize = 14)
plt.show()
```

```python
fig2 = plt.figure()
axis1 = fig2.add_axes([0, 0, 1, 1])
axis2 = fig2.add_axes([0.2, 0.5, 0.2, 0.2])

axis1.set_xlabel("X-Axis", fontsize = 12)
axis1.set_ylabel("Y-Axis", fontsize = 12)
axis1.set_title("Plot 2", fontsize = 14)
plt.show()
```

```python
fig3 = plt.figure()
axis1 = fig3.add_axes([0, 0, 1, 1])
axis2 = fig3.add_axes([0.2, 0.5, 0.4, 0.4])

axis1.set_xlabel("X-Axis", fontsize = 12)
axis1.set_ylabel("Y-Axis", fontsize = 12)
axis1.set_title("Plot 3", fontsize = 14)
plt.show()
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0, 0, 1, 1])
axis2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

x = range(10)
y = [i**2 for i in x]
z = [i**3 for i in x]

axis1.plot(x, z)
axis1.set_xlabel("X", fontsize = 12)
axis1.set_ylabel("Z", fontsize = 12)
axis1.set_title("X vs Z", fontsize = 14)

axis2.plot(x, y)
axis2.set_xlim([0, 5])
axis2.set_xlabel("X", fontsize = 12)
axis2.set_ylabel("Y", fontsize = 12)
axis2.set_title("X vs Y (X-limit set)", fontsize = 14)

plt.show()
```
