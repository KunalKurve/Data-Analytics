```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('alphabet_stock_data.csv')
df.info()
```

```python
# 1. Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.

df['Date'] = pd.to_datetime(df['Date'])

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# Defining the start and end dates
start_date = '2020-04-01'
end_date = '2020-05-01'

# Filter the DataFrame for the specified date range, loc used for labelled indexing
df_filtered = df.loc[start_date:end_date]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_filtered.index, df_filtered['Close'], color='green')
plt.title('Historical Stock Prices of Alphabet Inc. between ' + start_date + ' and ' + end_date)
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.grid(True)
plt.show()

df_filtered.info()
```

```python
# 2. Write a Pandas program to create a line plot of the opening, closing stock prices of Alphabet Inc. between two specific dates.

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_filtered.index, df_filtered['Open'], color='blue',label='Open')
plt.plot(df_filtered.index, df_filtered['Close'], color='red',label='Close')
plt.title('Opening and Closing Stock Prices of Alphabet Inc. between ' + start_date + ' and ' + end_date)
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Write a Pandas program to create a bar plot of the trading volume of Alphabet Inc. stock between two specific dates.

# Creating the bar plot
plt.figure(figsize=(14,7))
plt.bar(df_filtered.index, df_filtered['Volume'], color='blue')

plt.title('Alphabet Inc. Trading Volume Between April 1, 2020, and September 30, 2020')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=45) # Rotates the date labels
plt.tight_layout() # Adjusts the plot to ensure everything fits without overlap
plt.grid(True)
plt.show()
```

```python
# 4. Write a Pandas program to create a bar plot of opening, closing stock prices of Alphabet Inc. between two specific dates.

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df_filtered.index, df_filtered['Open'], color='blue',label='Open')
plt.bar(df_filtered.index, df_filtered['Close'], color='orange',label='Close')
plt.title('Opening and Closing Stock Prices of Alphabet Inc. between ' + start_date + ' and ' + end_date)
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 12. Write a Pandas program to create a plot of stock price and trading volume of Alphabet Inc. between two specific dates.

fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot closing price
ax1.set_xlabel('Date')
ax1.set_ylabel('Closing Price')
ax1.plot(df_filtered.index, df_filtered['Close'], color='blue')
ax1.tick_params(axis='y')

# Create a second y-axis for trading volume
ax2 = ax1.twinx()
ax2.set_ylabel('Volume')
ax2.bar(df_filtered.index, df_filtered['Volume'], color='green', alpha=0.3)
ax2.tick_params(axis='y')

fig.tight_layout()
plt.show()
```

```python
# 13. Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of Alphabet Inc. between two specific dates.

fig, axs = plt.subplots(6, figsize=(12, 18))

# Plot each column
columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
for i, column in enumerate(columns):
    axs[i].plot(df_filtered.index, df_filtered[column])
    axs[i].set_title(f'{column} of Alphabet Inc.')
    axs[i].set_ylabel(column)

plt.tight_layout()
plt.show()
```

```python
# 14. Write a Pandas program to create a plot of adjusted closing prices, thirty days and forty days simple moving average of Alphabet Inc. between two specific dates.

df.reset_index()

df_filtered = df.loc[start_date:end_date]

# Calculate 30-day and 40-day simple moving averages
df_filtered['30_SMA'] = df_filtered['Adj Close'].rolling(window=30).mean()
df_filtered['40_SMA'] = df_filtered['Adj Close'].rolling(window=40).mean()

fig, ax = plt.subplots(figsize=(12, 8))

# Plot adjusted closing prices and moving averages
ax.plot(df_filtered.index, df_filtered['Adj Close'], label='Adjusted Close')
ax.plot(df_filtered.index, df_filtered['30_SMA'], label='30-Day SMA')
ax.plot(df_filtered.index, df_filtered['40_SMA'], label='40-Day SMA')

ax.set_title('Adjusted Closing Prices and Simple Moving Averages of Alphabet Inc.')
ax.set_xlabel('Date')
ax.set_ylabel('Price')

ax.legend()
plt.tight_layout()
plt.show()
```

```python
# 15. Write a Pandas program to create a plot of adjusted closing prices, 30 days simple moving average and exponential moving average of Alphabet Inc. between two specific dates.


# Calculate 30-day simple moving average
df_filtered['30_SMA'] = df_filtered['Adj Close'].rolling(window=30).mean()

# Calculate 30-day exponential moving average
df_filtered['30_EMA'] = df_filtered['Adj Close'].ewm(span=30, adjust=False).mean()

# Create subplots
fig, ax = plt.subplots(figsize=(12, 8))

# Plot adjusted closing prices and moving averages
ax.plot(df_filtered.index, df_filtered['Adj Close'], label='Adjusted Close')
ax.plot(df_filtered.index, df_filtered['30_SMA'], label='30-Day SMA')
ax.plot(df_filtered.index, df_filtered['30_EMA'], label='30-Day EMA')

ax.set_title('Adjusted Closing Prices and Moving Averages of Alphabet Inc.')
ax.set_xlabel('Date')
ax.set_ylabel('Price')

ax.legend()
plt.tight_layout()
plt.show()
```

```python
# 16. Write a Pandas program to create a scatter plot of the trading volume/stock prices of Alphabet Inc. stock between two specific dates.

# Create scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(df_filtered['Volume'], df_filtered['Close'], alpha=0.5)
plt.title('Scatter Plot of Trading Volume vs Closing Price')
plt.xlabel('Trading Volume')
plt.ylabel('Closing Price')
plt.show()
```

```python
# 17. Write a Pandas program to create a plot to visualize daily percentage returns of Alphabet Inc. stock price between two specific dates.

# Calculate daily percentage returns
df_filtered['Daily Return'] = df_filtered['Adj Close'].pct_change()

# Create plot
plt.figure(figsize=(12, 8))
plt.plot(df_filtered.index, df_filtered['Daily Return'], label='Daily Return')
plt.title('Daily Percentage Returns of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.show()
```

```python
# 18. Write a Pandas program to plot the volatility over a period of time of Alphabet Inc. stock price between two specific dates.

# Calculate daily returns
df_filtered['Daily Return'] = df_filtered['Adj Close'].pct_change()

# Calculate volatility
df_filtered['Volatility'] = df_filtered['Daily Return'].rolling(window=21).std() * np.sqrt(21)

# Create plot
plt.figure(figsize=(12, 8))
plt.plot(df_filtered.index, df_filtered['Volatility'], label='Volatility')
plt.title('Volatility of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```

```python
# 19. Write a Pandas program to create a histogram to visualize daily return distribution of Alphabet Inc. stock price between two specific dates.

# Calculate daily returns
df_filtered['Daily Return'] = df_filtered['Adj Close'].pct_change()

# As there are null values, we have to drop those values from dataframe
# Drop NA values
df_filtered = df_filtered.dropna()

# Create histogram
plt.figure(figsize=(12, 8))
plt.hist(df_filtered['Daily Return'], bins=50)
plt.title('Histogram of Daily Return Distribution')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()
```
